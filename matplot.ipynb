```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Using scatter plots (points)
x_axis = np.random.random(100) * 100
y_axis = np.random.random(100) * 100

plt.scatter(x_axis, y_axis, c='000', s=10, marker='*')
```

```python
#Using line plots (lines connecting the points)
years = [2006 + x for x in range(16)]
weights = [80, 83, 84, 86, 82, 81, 79, 83, 80, 82, 80, 82, 83, 81, 80, 79]

plt.plot(years, weights, c='red', lw=1, linestyle='--') # using a string like "r--" you can configure the line
```

```python
#Using bar plot (bars)

x = ['C++', 'C#', 'Python', 'Java', 'Go']
y = [20, 50, 140, 1, 45]

plt.bar(x, y, color='green', align='edge', width=0.9)
```

```python
ages = np.random.normal(20, 1.5, 1000)

plt.hist(ages, color='purple', bins=20)
```

```python
x = ['C++', 'C#', 'Python', 'Java', 'Go']
votes = [50, 24, 14, 6, 17]
explodes = [0.3, 0, 0, 0, 0.1]
plt.pie(votes, labels=x, explode=explodes,
        autopct='%.2f%%', pctdistance=0.8)
```

```python
heights = np.random.normal(172, 8, 300)

plt.boxplot(heights)
```
